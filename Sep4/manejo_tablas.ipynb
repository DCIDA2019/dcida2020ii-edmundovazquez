Comenzamos extrayendo las tablas de nuestro interes. Primero la tabla de salarios minimos.

```python
import os
import requests
import pandas as pd

#primero extraemos la tebla de salarios minimos.
#para eso extraemos el url

url = "https://idconline.mx/laboral/salarios-minimos/salarios-minimos-1986-2007 "
html = requests.get(url).content 
```

Se nos descarga el archivo html

```python
#generamos una variable para guardar la tabla
file = 'salarios_minimos.csv' 

#extraemos la lista que nos interesa 
df_list = pd.read_html(html, header = 1) 
#header establece el inicio de la tabla

#con esto indicamos dónde la tabla va a terminar:
df = df_list[-1] 
#Y la guardamos con esto
df.to_csv('salarios_minimos.csv') 
```
