```python
#importamos librerias
import numpy as np
import matplotlib.pyplot as plt

l= 1
g= 9.81
k= (g/l)


def euler(N):
    x= np.zeros(100)
    w= np.zeros(100)
    a=0.0
    b=10.0
    x[0]=0.5 #angulo inicial medido en radianes
    #w= 0 #deribada de theta inicial
    t= np.linspace(0,10,N) # a t=0
    h= ((b-a)/N)
    
    for i in range (N-1):
        w[i+1]= w[i] - h*k*np.sin(x[i])
        x[i+1]= x[i] + h*w[i+1]

    plt.plot(t, x)
#    return(x)

#x1= euler(100)  #numero de particiones
%timeit euler(100)
```
