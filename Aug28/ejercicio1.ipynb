Codigo aprovechando los arreglos de Numpy

```python
#importamos librerias
import numpy as np
import matplotlib.pyplot as plt

x= np.zeros(100)
w= np.zeros(100)

N= 100 #numero de particiones
t= np.linspace(0,10,N) # a t=0
l= 1
g= 9.81
k= (g/l)
x[0]=0.5 #angulo inicial medido en radianes
#w= 0 #deribada de theta inicial
a=0.0
b=10.0
h= ((b-a)/N)

for i in range (N-1):
    w[i+1]= w[i] - h*k*np.sin(x[i])
    x[i+1]= x[i] + h*w[i+1]

plt.plot(t, x)
plt.show()
```
